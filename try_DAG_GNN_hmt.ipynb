```python
def preprocess_adj_new(adj):
    adj_normalized = (torch.eye(adj.shape[0]).double() - (adj.transpose(0,1)))
    return adj_normalized

def preprocess_adj_new1(adj):
    adj_normalized = torch.inverse(torch.eye(adj.shape[0]).double()-adj.transpose(0,1))
    return adj_normalized

def isnan(x):
    return x!=x

def my_normalize(z):
    znor = torch.zeros(z.size()).double()
    for i in range(z.size(0)):
        testnorm = torch.norm(z[i,:,:], dim=0)
        znor[i,:,:] = z[i,:,:]/testnorm
    znor[isnan(znor)] = 0.0
    return znor
```

```python


class DAGNN(nn.Module):
    def __init__(self, n_nodes, n_in_features=1, n_hid_features=64, n_out_features=1, tau=0.1):
        super(DAGNN, self).__init__()
        
        self.n_nodes = n_nodes
        self.tau = tau
        
        # Initialize adjacency matrix
        self.adj_A = nn.Parameter(torch.zeros(n_nodes, n_nodes).double())
        
        # Encoder
        self.encoder = MLPEncoder(n_nodes * n_in_features, n_in_features, n_hid_features, n_out_features, self.adj_A)
        
        # Decoder
        self.decoder = MLPDecoder(n_nodes * n_in_features, n_out_features, n_in_features, self.encoder, n_nodes, batch_size=100, n_hid=n_hid_features)
        
        self.Wa = nn.Parameter(torch.zeros(n_out_features).double())
        self.z = nn.Parameter(torch.tensor(0.1).double())
        self.z_positive = nn.Parameter(torch.ones_like(self.adj_A).double())

    def forward(self, inputs):
        # Prepare relation matrices (assuming fully connected for simplicity)
        rel_rec = rel_send = torch.ones(self.n_nodes, self.n_nodes).double()
        
        # Encoder
        encoded, logits, adj_A1, adj_A, z, z_positive, adj_A_raw, Wa = self.encoder(inputs, rel_rec, rel_send)
        
        # Decoder
        mat_z, output, adj_A_tilt = self.decoder(inputs, encoded, self.n_nodes, rel_rec, rel_send, adj_A1, adj_A, Wa)
        
        return output, adj_A_tilt, adj_A1, self.adj_A, z, z_positive, encoded, logits

class MLPEncoder(nn.Module):
    def __init__(self, n_in, n_xdims, n_hid, n_out, adj_A):
        super(MLPEncoder, self).__init__()

        self.adj_A = adj_A
        self.fc1 = nn.Linear(n_xdims, n_hid)
        self.fc2 = nn.Linear(n_hid, n_out)
        self.Wa = nn.Parameter(torch.zeros(n_out).double())

    def forward(self, inputs, rel_rec, rel_send):
        # Preprocessing adjacency matrix
        adj_A1 = torch.sinh(3. * self.adj_A)
        adj_Aforz = preprocess_adj_new(adj_A1)
        adj_A = torch.eye(adj_A1.size()[0]).double()

        H1 = F.relu(self.fc1(inputs))
        x = self.fc2(H1)
        logits = torch.matmul(adj_Aforz, x + self.Wa) - self.Wa

        return x, logits, adj_A1, adj_A, None, None, self.adj_A, self.Wa

class MLPDecoder(nn.Module):
    def __init__(self, n_in_node, n_in_z, n_out, encoder, data_variable_size, batch_size, n_hid):
        super(MLPDecoder, self).__init__()

        self.out_fc1 = nn.Linear(n_in_z, n_hid)
        self.out_fc2 = nn.Linear(n_hid, n_out)
        self.batch_size = batch_size
        self.data_variable_size = data_variable_size

    def forward(self, inputs, input_z, n_in_node, rel_rec, rel_send, origin_A, adj_A_tilt, Wa):
        adj_A_new1 = preprocess_adj_new1(origin_A)
        mat_z = torch.matmul(adj_A_new1, input_z + Wa) - Wa

        H3 = F.relu(self.out_fc1(mat_z))
        out = self.out_fc2(H3)

        return mat_z, out, adj_A_tilt

def train_dagnn(model, optimizer, data_loader, num_epochs=300):
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0
        for batch_idx, data in enumerate(data_loader):
            optimizer.zero_grad()
            print(data)
            output, adj_A_tilt, adj_A1, adj_A, z, z_positive, encoded, logits = model(data)
            
            # Calculate loss (you need to define your loss function)
            loss = calculate_loss(output, data, adj_A, z)
            
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
        
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {total_loss/len(data_loader):.4f}')

def calculate_loss(output, target, adj_A, z):
    # Define your loss function here
    # This is a placeholder and needs to be replaced with your actual loss calculation
    mse_loss = F.mse_loss(output, target)
    dag_constraint = torch.trace(torch.matrix_exp(adj_A * adj_A)) - adj_A.size(0)
    return mse_loss + 0.1 * dag_constraint + 0.01 * torch.sum(torch.abs(adj_A))

# Usage example
n_nodes = 10  # number of variables
model = DAGNN(n_nodes, n_in_features=n_nodes)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader

class DAGDataset(Dataset):
    def __init__(self, dataframe):
        self.data = torch.tensor(dataframe.values, dtype=torch.float32)
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return self.data[idx]

def create_train_loader(dataframe, batch_size=100, shuffle=True):
    dataset = DAGDataset(dataframe)
    return DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

df = X_train['00000']

# 使用示例：
# 假设您的 DataFrame 名为 df
train_loader = create_train_loader(df)
```
